```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up the driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Navigate to the login page
login_url = "https://central1.recallmax.com/centralServer/request/admin/public/user/login.html?redirectUrl=https%3A%2F%2Freportcan.recallmax.com%2Freport%2Frequest%2Fsecure%2Fdash%2FlistReport%2FlistReport.html%3FaccountId%3D71475%26period%3D2024-09&redirectServer=https%3A%2F%2Freportcan.recallmax.com&redirectContext=%2Freport"
driver.get(login_url)

# Give the page time to load
time.sleep(2)

# Input username and password
username_input = driver.find_element(By.NAME, "username")
password_input = driver.find_element(By.NAME, "password")

username_input.send_keys("JeremyBarr")
password_input.send_keys("auKYNt5njZgrNHeF")

# Submit the form
password_input.send_keys(Keys.RETURN)

# Wait for the page to load after login
time.sleep(5)

# Navigate to the survey page
survey_url = "https://can5.recallmax.com/rsm/request/dash/secure/survey/surveyResponses.html?accountId=67640&period=2024-09"
driver.get(survey_url)

# Wait for the survey page to load
time.sleep(5)

# Scrape the survey table rows
survey_rows = driver.find_elements(By.CSS_SELECTOR, "table tr")  # Adjust based on table structure

print(f"\nNumber of rows found: {len(survey_rows)}\n")

# Loop through each row in the table
for i in range(1, len(survey_rows)):
    try:
        # Re-find all rows (since the page reloads after each click, we need to find elements again)
        survey_rows = driver.find_elements(By.CSS_SELECTOR, "table tr")
        
        # Click the current row (adjust index based on table structure, skipping the header)
        row = survey_rows[i]
        row.click()

        # Wait for the page to load patient details
        time.sleep(5)

        # Scrape the patient details (adjust the selector based on the details page structure)
        patient_details = driver.find_elements(By.CSS_SELECTOR, "div.patient-details")  # Adjust based on actual structure
        
        print(f"\nPatient Details for row {i}:\n")
        for detail in patient_details:
            print(detail.text)  # Print or save the patient details

        # After scraping, navigate back to the survey table
        driver.back()

        # Wait for the table to load again
        time.sleep(5)

    except Exception as e:
        print(f"Error processing row {i}: {e}")
```
